```python
from orbital_simulation import Simulation, Rigidbody

# Simulate one day per step
sim = Simulation(dt=86400)

# load all planets up to mars and don't include the moon.
sim.load_planets(cutoff_index=6, include_moon=False)

# Simulate for 1.8 years (approx. period of Mars)
sim.run(time=86400 * 365 * 1.88)

fig, ax = sim.animate(
    steps_per_frame=2,
    framesep=30,
    save_file="inner_solar_system.gif",
    center_view=True,
)
fig, ax = sim.animate(
    steps_per_frame=2,
    framesep=30,
    save_file="inner_solar_system_relative.mp4",
    center_view=True,
    center_body=sim.get_rigidbody_by_name("earth"),
)
```
